### 3) Use of Anonymous functions with Built-In functions
#### 3.1) lambda function with filter - Program to filter odd numbers from list of numbers

```python
my_list = [1, 2, 12, 34, 65, 67, 35, 72]
filtered_list = list(filter(lambda x: x%2 != 0, my_list))
filtered_list
```

#### 3.2) lambda() function with map() - Program to multiply each element of the list with 2

```python
my_list = [5, 10, 12, 3, 13, 4, 25]
mapped_list = list(map(lambda x: x*2, my_list))
mapped_list
```

### 4) Return Statement

```python
def sum_two_numbers(num1, num2):
    'This function returns sum of two numbers'
    
    total = num1 + num2
    return total

total = sum_two_numbers(2, 5)
print('Sum is: ', total)
```

### 5) Recursive Function

```python
def factorial(n):
    '''This function finds the
        factorial of a number'''
    
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)
    
ans = factorial(5)
print('Factorial is: ', ans)
```

### 6) Scope of Variables

```python
total = 0   # Global variable

def find_sum(num1, num2):
    'This function finds sum of two numbers'
    
    total = num1 + num2  # Here total is Local Variable
    print('Print total inside the function: ', total)
    return total

find_sum(1, 2)
print('Print total outside the function: ', total)
```

Example 2

```python
global_string = 'Hello World!'

def print_string():
    '''This function will print Global and 
        Local Variables'''
    
    print('Printing Global Variable inside function: ', global_string)
    local_string = 'This is Local Variable'
    print('Printing Local Variable inside function: ', local_string)
    return None

print_string()
print('----------')
print('Printing Global Variable Outside function: ', global_string)
print('Printing Local Variable Outside function: ', local_string)
```

<b>Ques 1</b>: Program to mask an image

```python
# Loading data from skimage.data
image = data.camera()

# Creating figure
f = plt.figure(figsize=(10,25))
# Creating subplot and ploting the image
ax1 = f.add_subplot(121)
ax1.set_title('Original Image')
ax1.imshow(image)


# Creating mask for the image
mask = image < 87
image[mask] = 255

ax2 = f.add_subplot(122)
ax2.set_title('Masked Image')
ax2.imshow(image, cmap='gray')
```
